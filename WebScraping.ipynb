```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}
url = "https://www.flipkart.com/search?q=mobile%20phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
url
```

```python
r = requests.get(url)
r
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
container = soup.find_all("div", class_="cPHDOP col-12-12")
len(container)
```

```python
names=[]
ratings=[]
price=[]
```

```python
for i in range(len(container)):
    names.append(container[i].find('div', class_='KzDlHZ'))
    try:
        ratings.append(container[i].find('div', class_='XQDdHH'))
    except:
        ratings.append(None)
    price.append(container[i].find('div', class_='Nx9bqj _4b5DiR'))
```

```python
data={'Name':names, 'Ratings':ratings, 'Price':price}
df = pd.DataFrame(data)
df
```

```python
url2 = "https://www.flipkart.com/samsung-galaxy-f14-5g-goat-green-128-gb/product-reviews/itm032d1a69999cc?pid=MOBGNBFNDPGNJ7HY&lid=LSTMOBGNBFNDPGNJ7HYRCCFUH&marketplace=FLIPKART"
url2
```

```python
r2 = requests.get(url2)
r2
```

```python
soup2 = BeautifulSoup(r2.text, "lxml")
```

```python
container2 = soup2.find_all("div", class_="cPHDOP col-12-12")
len(container2)
```

```python
reviewsss=[]
commentsss=[]
cust_namess=[]
cust_ratingss=[]
```

```python
for i in range(len(container2)):
    try:
        reviewsss.append(container2[i].find_all('p', class_="z9E0IG"))
    except:
        reviewsss.append(None)
    try:
        cust_namess.append(container2[i].find_all('p', class_="_2NsDsF AwS1CA"))
    except:
        cust_namess.append(None)
    try:
        cust_ratingss.append(container2[i].find_all('div', class_="XQDdHH Ga3i8K").text)
    except:
        cust_ratingss.append(None)
    try:
        commentsss.append(container2[i].find_all('div', class_="ZmyHeo").text)
    except:
        commentsss.append(None)
```

```python
table_data = {"Customer Name":cust_namess, "Review":reviewsss,"Comment":commentsss,"Rating":cust_ratingss}
data10 = pd.DataFrame(table_data)
data10
```
