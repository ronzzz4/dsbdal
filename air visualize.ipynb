```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('airqualityuci.csv')
df.head()
```

```python
sns.barplot(x='Time',y='NMHC(GT)',data=df)
```

```python
sns.scatterplot(x='NO2(GT)', y='NMHC(GT)', data=df)
```

```python
sns.lineplot(x='Time',y='NO2(GT)',data=df)
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(df,hue="Time", palette="Blues")
```

```python
sns.displot(df['NO2(GT)'])
```

```python
timedf = pd.DataFrame(df['Time'])
timedf
```

```python
plt.figure(figsize=(6,6))
plt.pie(
    timedf['Time'].value_counts(),
    labels = timedf['Time'].value_counts().index
)
plt.title('Time distribution')
plt.show()
```
