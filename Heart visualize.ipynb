```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('heartdisease.csv')
```

```python
df.head()
```

```python
sns.barplot(x='sex', y='thalach', data=df)
```

```python
sns.scatterplot(x='age', y='thalach', data=df)
```

```python
sns.lineplot(x='sex',y='age',data=df)
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(df, hue='sex', palette='Blues')
```

```python
sns.displot(df['thalach'])
```

```python
# Scatter plot using matplotlib
plt.scatter(df['thalach'], df['age'])
plt.show()
```

```python
sexdf = pd.DataFrame(df['sex'])
sexdf
```

```python
plt.figure(figsize=(6,6))
plt.pie(sexdf['sex'].value_counts(), labels=sexdf['sex'].value_counts().index)
plt.title('Distribution of Sex')
# plt.axis('equal')
plt.show()
```

```python
plt.subplot(1, 2, 2)
plt.bar(sexdf['sex'].value_counts().index, sexdf['sex'].value_counts())
plt.title('Bar plot - Distribution of sex')
plt.xlabel('Sex')
plt.ylabel('Count')
# plt.tight_layout()
plt.show()
```
